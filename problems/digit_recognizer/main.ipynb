```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
```

```python
train_data = pd.read_csv('train.csv')
```

```python
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
y_train = train_data['label']
x_train = train_data.drop(['label'], axis=1)
```

```python
y_train.head()
```

```python
x_train.head()
```

```python
x_train /= 255
```

```python
clf = MLPClassifier(solver="sgd", alpha=1e-5, max_iter=80,
                    hidden_layer_sizes=(40,), verbose=10,
                    learning_rate_init=0.2, random_state=1)
```

```python
clf.fit(x_train, y_train)
```

```python
x_test = pd.read_csv('test.csv')
```

```python
x_test.head()
```

```python
x_test.isnull().sum()
```

```python
x_test /= 255
```

```python
y_test = clf.predict(x_test)
y_test[:5]
```

```python
!touch submission.csv
```

```python
with open("submission.csv", "w") as file:
    file.writelines("ImageId,Label\n")
    for i in range(0, len(y_test)):
        file.writelines(f'{i+1},{y_test[i]}\n')
```
